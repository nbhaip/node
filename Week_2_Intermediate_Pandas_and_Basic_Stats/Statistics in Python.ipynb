# Parametric Statistics
This is a group of statistics that is focused on normal distribution over a population or large sample set.  If you have ever worked in a biology lab, there is a good chance you have worked with it.  To properly analyze a data set with parametric statistics 

## Hypotheses
There are two kinds of hypotheses for a one sample t-test, the null hypothesis and the alternative hypothesis. The alternative hypothesis assumes some difference between the true mean value and the comparison value, whereas the null hypothesis assumes that no difference exists. The purpose of the one sample t-test is to determine if the null hypothesis should be rejected, given the sample data. 

## T-Test
T-Tests are an important way to analyze the variance between two populations.  
### 1 Sample T-Test
These tests can be used to analyze variance within a large dataset. Effectively, this shows how much variance exists within a dataset. 

```python
stats.ttest_1samp(data['Rate'], 0)
```

### 2 Sample T-Test
This is a great way to find significant differences between two specific samples within one dataset.  As seen in the example below, these t-tests are useful for users to be able to discern significant difference between two subsets within these datasets. These are different from-paired t-tests that test various hypothesises, with different independent variables. Paired t-tests therefore allow for the user to be able to remove inter-subject variability. 

```python
# Two Sample T-Tests
male_rate = data[data['Gender'] == '2']['Temp']
female_rate = data[data['Gender'] == '1']['Temp']
stats.ttest_ind(male_rate, female_rate) 
```

```python
# Paired T-Tests
stats.ttest_ind(data['Rate'], data['Temp'])

#
stats.ttest_rel(data['Rate'], data['Temp'])
```

# Non Parametric Statistics

In fact, one is more likely to run into a situation where one will need non parametric statistics, which for the most part returns extremely similar results as does most parametric statistics analysis. 
## Mann_Whitney Test
The Non-Parametric Version of the t-test, the Mann-Whitney Test offers the user to apply the same T-Test for Significance onto non-parametric data.
